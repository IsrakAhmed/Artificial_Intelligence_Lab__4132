# Prepare a neural network with appropriate activation functions, so that it has the same architecture as shown in the following Figure-2.

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, GlobalAveragePooling2D
```

```python
inputs = Input((256, 256, 3), name='Input_Layer')

x = Conv2D(64, (3, 3), activation='relu', padding='same', name='Conv2D_1')(inputs)
x = MaxPooling2D((2, 2), name='MaxPooling2D_1')(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same', name='Conv2D_2')(x)

x = GlobalAveragePooling2D(name='GlobalAveragePooling2D')(x)

outputs = Dense(1, activation='sigmoid', name='Output_Layer')(x)

model = Model(inputs, outputs, name='My_Model')

model.summary()
```
